```python
#A partir d’un jeu de données, représentez un nuage de points basé sur deux variables continues

import pandas as pd
import xlrd, openpyxl

dataxls=pd.ExcelFile("iris.xlsx")
iris=dataxls.parse('Feuil1')

plt.plot(iris['Long. Sepales'],iris['Long. Petales'],'bo')
plt.title('Les iris')
plt.xlabel('Long. Sepales')
plt.ylabel('Long. Petales')
```

```python
#Toujours sur les iris, représenter un scatter plot avec une couleur par espèce pour les deux premières variables


iris.groupby("Espece").describe()
iris_grouper=iris.groupby("Espece")

colors = ['r', 'g' , 'b'] #we'll be cycling through these colors
color_index = 0
for group in iris_grouper:
    color = colors[color_index]
    group_label = group[0]
    group_data = group[1]
    plt.plot(group_data['Long. Sepales'],group_data['Long. Petales'], "o",color=color, label=group_label)
    color_index += 1
plt.legend(loc=0)
```

```python
#Générez un histogramme à partir de 2 échantillons de 100 observations sur 20 intervalles
donnees1 = np.random.randn(100)
donnees2= np.random.randn(100)

plt.hist(donnees1,normed=True,color='r',alpha=0.5,bins=20,label='group1')
plt.hist(donnees2,normed=True,alpha=0.5,bins=20, label='group2')
plt.legend()
```
